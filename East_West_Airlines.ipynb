# 1. Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering, KMeans, DBSCAN

import warnings
warnings.filterwarnings('ignore')
```

```python
airlines = pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
airlines
```

### No null entries in the dataset

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

# Clustering

## A. Hierarchical Clustering

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(airlines.iloc[:,1:])
df_norm
```

### Create Dendogram

```python
plt.figure(figsize=(15,10))
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'single')
```

```python
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
airlines['Hierarchical clustering']=y_hc
airlines
```

## B. Kmeans Clustering

```python
k=list(range(1,11))
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 5))
wcss = []
for v in k:
    clf = KMeans(n_clusters=v)
    clf.fit(df_norm)
    wcss.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.ylabel('Inertia')
plt.xlabel('Values of k')
plt.show()
```

### To determine the optimal number of clusters, we have to select the value of k at the “elbow” ie the point after which the distortion/inertia start decreasing in a linear fashion. Thus for the given data, we can conclude that the optimal number of clusters for the data is 5

## Building Kmeans clustering algorithm

```python
clusters_new = KMeans(n_clusters=5, random_state=12)
clusters_new.fit(df_norm)
```

```python
clusters_new.labels_
```

```python
airlines['Kmeans']=clusters_new.labels_
airlines
```

```python
kmeans=airlines['Kmeans']
```

```python
airlines.plot(x="Balance",y ="Qual_miles",c=kmeans ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using KMeans') 
```

## Silhouette score

```python
from sklearn import metrics
```

```python
metrics.silhouette_score(df_norm, kmeans)
```

## C. DBSCAN

```python
df = pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
```

```python
df1 = df.drop(['ID#'],axis=1)
```

```python
df1
```

```python
array = df1.values
array
```

```python
std = StandardScaler()
X = std.fit_transform(array) 
X 
```

```python
dbscan = DBSCAN(eps=0.95, min_samples=5)
dbscan.fit(X)
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster']) 
```

```python
cl
```
